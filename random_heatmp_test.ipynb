```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

image = np.random.rand(3,3)

# Create the heatmap
fig = plt.figure(0)
plt.imshow(image, cmap='viridis', interpolation='nearest')

# Add a colorbar for reference
plt.colorbar()

# Add labels (optional)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Heatmap Pre weights ReLU-ReLU')

# Show the plot
plt.show()
print(image)
```

```python
example = np.array([[0.96503288, 0.45242525, 0.30052466], [0.65297134, 0.11945828, 0.93953196], [0.22197709, 0.80758823, 0.82971406]])

print("BASE EXAMPLE")
print(np.min(example))
print(np.median(example))
print(np.max(example))
print(np.mean(example))
print(np.std(example))
print(np.linalg.det(example))
print(np.linalg.eigvals(example))

example = np.matrix(example)

mod_example = example - 0.5
print("MOD EXAMPLE")
print(np.linalg.det(mod_example))
print(np.linalg.eigvals(mod_example))
```

```python
# Create the heatmap
fig = plt.figure(0)
plt.imshow(mod_example, cmap='viridis', interpolation='nearest')

# Add a colorbar for reference
plt.colorbar()

# Add labels (optional)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Heatmap Pre weights ReLU-ReLU')

# Show the plot
plt.show()
print(mod_example)
```

```python
import random

def sigmoid(arr):
    return 1 / (1 + np.exp(-1*arr))

def relu(arr):
    return np.maximum(arr, 0)

def swish(arr):
    return np.multiply(arr,sigmoid(arr))

def elu(arr):
    new_arr = np.ones(arr.shape)
    for index, item in enumerate(arr):
        if item > 0:
            new_arr[index] = item
        else:
            new_arr[index] = np.exp(item)-1
    return new_arr

test_input = np.random.rand(3)
print(test_input.shape)
bias = random.random()
compute_1 = test_input*example + bias
compute_2 = test_input*mod_example + bias
compute_1 = np.asarray(compute_1).reshape(3)
compute_2 = np.asarray(compute_2).reshape(3)
example_mult = elu(compute_1)
mod_example_mult = elu(compute_2)
print(example_mult)
print(mod_example_mult)
print(example_mult - mod_example_mult)
```
