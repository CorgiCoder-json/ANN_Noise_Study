```python
example = np.array([[-1, 0.45242525, 0.30052466], [0.65297134, -1, 0.93953196], [0.22197709, 0.80758823, -1]])

print("BASE EXAMPLE")
print(np.min(example))
print(np.median(example))
print(np.max(example))
print(np.mean(example))
print(np.std(example))
print(np.linalg.det(example))
print(np.linalg.eigvals(example))

example = np.matrix(example)
mod_mat = np.random.rand(3, 3)
mod_example = example * .3 + mod_mat * .7
print("MOD EXAMPLE")
print(np.linalg.det(mod_example))
print(np.linalg.eigvals(mod_example))

print("EIGENVAL Change")
print(np.linalg.eigvals(example)/np.linalg.eigvals(mod_example))
```

```python
import random

def sigmoid(arr):
    return 1 / (1 + np.exp(-1*arr))

def relu(arr):
    return np.maximum(arr, 0)

def swish(arr):
    return np.multiply(arr,sigmoid(arr))

def sq(arr):
    return np.multiply(arr, arr)

def elu(arr):
    new_arr = np.ones(arr.shape)
    for index, item in enumerate(arr):
        if item > 0:
            new_arr[index] = item
        else:
            new_arr[index] = np.exp(item)-1
    return new_arr

test_input = np.random.rand(3)
bias = np.random.rand(3)
print(test_input)
print(bias)
compute_1 = test_input*example + bias
compute_2 = test_input*mod_example + bias
compute_1 = np.asarray(compute_1).reshape(3)
compute_2 = np.asarray(compute_2).reshape(3)
example_mult = relu(compute_1)
mod_example_mult = relu(compute_2)
print(example_mult)
print(mod_example_mult)
print(example_mult - mod_example_mult)
```

```python
rand_mat = np.random.rand(3,3)
rand_vec = np.random.rand(3,)
bias_vec = np.random.rand(3,)
print(f"WEIGHT: {rand_mat}")
print(f"INPUT: {rand_vec}")
print(f"BIAS: {bias_vec}")
inv_mat = np.linalg.inv(rand_mat)
print(f"INVERSE: {inv_mat}")
result = rand_vec.dot(rand_mat) + bias_vec
print(f"FORWARD RESULT: {result}")
inv_result = (result-bias_vec).dot(inv_mat)
print(f"INVERSE RESULT: {inv_result}")
```

```python
rand_mat = np.random.rand(2,4)
rand_vec = np.random.rand(2,)
bias_vec = np.random.rand(4,)
rand_add = np.random.rand(4,)
print(f"WEIGHT: {rand_mat}")
print(f"INPUT: {rand_vec}")
print(f"BIAS: {bias_vec}")
result = rand_vec.dot(rand_mat) + bias_vec
print(f"FORWARD RESULT: {result}")
app_mat = np.append(rand_mat, np.array([rand_add, np.random.rand(4,)]), 0)
inv_mat = np.linalg.inv(app_mat)
inv_result = (result-bias_vec).dot(inv_mat)
print(f"INVERSE RESULT: {inv_result}")
rand_mat = rand_mat + np.random.rand(4,)
print(f"NEW WEIGHTS: {rand_mat}")
result = rand_vec.dot(rand_mat) + bias_vec
print(f"FORWARD 2: {result}")
app_mat = np.append(rand_mat, np.array([rand_add, np.random.rand(4,)]), 0)
inv_mat = np.linalg.inv(app_mat)
inv_result = (result-bias_vec).dot(inv_mat)
print(f"INVERSE RESULT 2: {inv_result}")
```
